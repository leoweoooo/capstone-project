```python
import pandas as pd

cnn_shap = pd.read_excel('./data/cnn/shap.xlsx', index_col=0)
mlp_shap = pd.read_excel('./data/ffn/shap.xlsx', index_col=0)
```

```python
full_df = pd.concat([cnn_shap, mlp_shap], axis=0)
full_df = full_df.groupby(full_df.index).mean()
full_df = full_df[~(full_df == 0).all(axis=1)]
full_df.to_csv('./data/sigs.csv')
full_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 8))
sns.heatmap(full_df, cmap='RdBu_r')
plt.xlabel('Species and Environment')
plt.ylabel('mVOCs')
plt.savefig('./data/figures/mean-shap.png', dpi=300, bbox_inches='tight')
plt.show()
```
