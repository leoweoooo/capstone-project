```python
from modules import data 

records = data.from_json('./data/json/ind.json')
[record for record in records if (record.trial == 1 or record.trial == 2)]
for record in records:
    record.min_quality(90)
    
records_df = data.to_df(records)
records_df
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
records_df['Label'] = label_encoder.fit_transform(records_df['Label'])
```

```python
rec_df = records_df.set_index('Label')
rec_df = data.standard_scaler(rec_df)
rec_df
```

```python
# Separation to Train and Test
from sklearn.model_selection import train_test_split
train_datas, test_datas, train_labels, test_labels = train_test_split(rec_df.values, rec_df.index, train_size=0.2, random_state=42, stratify=rec_df.index)
```

## Wrapper Methods

### Support Vector Machine - Recursive Feature Elimination (SVM-RFE)

The accuracy (F1 Metric) of the fitted model when tested appears to be around 9-11%. There is a slight increase when reducing the amount of information (i.e., passing in `Ae` records only) but it is not significant enough for me to consider it a major win. 

It is likely that the information from this model is not an accurate representation. 

```python
# Feature Selection with SVM-RFE
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.metrics import f1_score, roc_auc_score

svm_est = SVC(kernel='linear')
svm_rfe = RFE(estimator=svm_est, n_features_to_select=50)
svm_rfe.fit(train_datas, train_labels)

f1_score(svm_rfe.predict(test_datas), test_labels, average='weighted')
```

## Embedded Methods

### Random Forest Classifier

The accuracy of this model is higher than SVM-RFE but does not happen to be significantly higher. It still tends to dwindle below 50%. 

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import f1_score, roc_auc_score

rf_clf = RandomForestClassifier(n_estimators=100, criterion='gini')
rf_clf.fit(train_datas, train_labels)

f1_score(rf_clf.predict(test_datas), test_labels, average='weighted')
```
